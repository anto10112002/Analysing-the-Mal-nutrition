```python
#4th dataset
df = pd.read_csv("country-wise-average.csv")
df
```

<span style="font-family:cursive; font-size:25px; color:green;"> Stunting</span>


```python
import plotly.express as px
name = df.groupby("Country")["Stunting"].mean().sort_values(ascending=False).index[:50]
yax = df.groupby("Country")["Stunting"].mean().sort_values(ascending=False).round()[:50]

fig = px.bar(df, y=yax, x=name, color = name, color_discrete_sequence=px.colors.sequential.Blugrn)
fig.update_layout(
    title="Stunting Top 50 Countries",
    xaxis_title="Country name",
    yaxis_title="Count"
)
fig.update_xaxes(tickangle=-45)

fig.show()
```

```python
import plotly.graph_objs as go
x = df.groupby(["Country"])["Stunting"].mean()
layout = dict(geo = {'scope':'world'},title="Stunting percentage")
map = go.Figure(data = [dict(type = 'choropleth',
            locations = x.index,
            locationmode = 'country names',
            colorscale= px.colors.sequential.Blugrn,
            text= x.index,
            z=x,
            colorbar = {'title':'stunting %' })],layout = layout)
map.show()
```

<span style="font-family:cursive; font-size:25px; color:green;"> Top 10 Country with Childhood Malnutrition</span>

Lastly we will see which 10 countries which has highest average of four symptoms

```python
df['Average'] = df.Overweight + df.Underweight + df.Wasting + df.Stunting / 4
```

```python
name3 = df.groupby("Country")["Average"].mean().sort_values(ascending=False).index[:10]
yax3 = df.groupby("Country")["Average"].mean().sort_values(ascending=False).round()[:10]

fig = px.bar(df, y=yax3, x=name3, color = name3, color_discrete_sequence=px.colors.sequential.Tealgrn)
fig.update_layout(
    title="Top 10 Countries",
    xaxis_title="Country name",
    yaxis_title="Count"
)
fig.update_xaxes(tickangle=-45)

fig.show()
```

```python
x = df.groupby(["Country"])["Average"].mean()
layout = dict(geo = {'scope':'world'},title="Overall Percentage")
map = go.Figure(data = [dict(type = 'choropleth',
            locations = x.index,
            locationmode = 'country names',
            colorscale= px.colors.sequential.Viridis,
            text= x.index,
                              
            z=x,
            colorbar = {'title':'stunting %' })],layout = layout)
map.show()
```
